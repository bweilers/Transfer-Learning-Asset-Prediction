```python
import time
import datetime
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# import hvplot.pandas
%matplotlib inline

import yfinance as yf
from pandas_datareader import data as pdr
```

```python
"""
This function gets and renames the columns form yfinance
"""
def get_named_data(ticker):
  data = yf.download(ticker, period = "max")
  data.columns = [f'{ticker}_Open', f'{ticker}_High', f'{ticker}_Low', f'{ticker}_Close', f'{ticker}_Adj_Close', f'{ticker}_Volume']
  return data
```

```python
ftt_data = get_named_data("FTT-USD")
# Bnb
bnb_data = get_named_data("BNB-USD")
# HT
ht_data = get_named_data("HT-USD")
# OKB
okb_data = get_named_data("OKB-USD")
# KCS
kcs_data = get_named_data("KCS-USD")
# LEO
leo_data = get_named_data("LEO-USD")

# concat
combined_crypto = pd.concat([ftt_data,bnb_data,ht_data,okb_data,kcs_data,leo_data], axis=1)
combined_crypto
```

```python
# Assess null
combined_crypto.isnull().mean() * 100
```

```python
# Drop null and review
combined_crypto = combined_crypto.dropna()
display(combined_crypto.shape)
combined_crypto.isnull().mean() * 100
```

```python
# Remove the hour to make the datetime index compatible with the other dataframes for concatinating
combined_crypto.index = combined_crypto.index.date

combined_crypto
```

```python
#  Get the data of possibly correlated stocks
square_data = get_named_data_FEATURE("SQ")
paypal_data = get_named_data_FEATURE("PYPL")
tesla_data = get_named_data_FEATURE("TSLA")
microstrategy_data = get_named_data_FEATURE("MSTR")
BNY_data = get_named_data_FEATURE("BK")

# combine
combined_stock_data = pd.concat([square_data,paypal_data,tesla_data,microstrategy_data,BNY_data], axis=1)
combined_stock_data
```

```python
# Remove the hour to make the datetime index compatible with the other dataframes for concatinating
combined_stock_data.index = combined_stock_data.index.date

combined_stock_data
```

```python
# Assess null
combined_stock_data.isnull().mean() * 100
```

```python
# Drop null and review
combined_stock_data = combined_stock_data.dropna()
display(combined_stock_data.shape)
combined_stock_data.isnull().mean() * 100
```
