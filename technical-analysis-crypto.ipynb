```python
def create_df_with_technical_analysis(token_ticker):
  # Create df for single token
  TOKEN_df = crypto_df.filter(regex=token_ticker)

  #NEW Rename for new FINTA TA
  TOKEN_df = TOKEN_df.rename(columns = {f'{token_ticker}-USD_Close':'Close', f'{token_ticker}-USD_High':'High', f'{token_ticker}-USD_Low':'Low', f'{token_ticker}-USD_Open':'Open', f'{token_ticker}-USD_Volume':'Volume', f'{token_ticker}_DailyReturn':'Returns'})

  # pull feature columns and concatinate to df
  features = crypto_df.filter(regex='FEATURE')
  TOKEN_df = pd.concat([TOKEN_df, features], axis = 1)

  # TOKEN_df.head()

  # Set SMA long and short for token
  short_window = 1
  long_window = 5

  # # Create SMAs that use short_window, long_window, and assign to new columns 
  # TOKEN_df['sma_fast'] = TOKEN_df[f'{token_ticker}-USD_Close'].rolling(window=short_window).mean()
  # TOKEN_df['sma_slow'] = TOKEN_df[f'{token_ticker}-USD_Close'].rolling(window=long_window).mean()
  # Create SMAs that use short_window, long_window, and assign to new columns 
  TOKEN_df['sma_fast'] = TOKEN_df[f'Close'].rolling(window=short_window).mean()
  TOKEN_df['sma_slow'] = TOKEN_df[f'Close'].rolling(window=long_window).mean()


  # Drop the NaNs using dropna()
  TOKEN_df = TOKEN_df.dropna()

  # preview
  # FTT_df.head()

  # Create a new column in the `trading_df` called "signal" setting its value to zero.
  TOKEN_df["Signal"] = 0.0

  # Create the signal to buy and sell
  TOKEN_df.loc[(TOKEN_df["sma_fast"] > TOKEN_df["sma_slow"]), "Signal"] = 1
  TOKEN_df.loc[(TOKEN_df["sma_fast"] < TOKEN_df["sma_slow"]), "Signal"] = -1

  # add in more TA with FINTO
  # List of time periods to use for Moving Averages calculation
  timeperiods = [15,60]

  # create UCO TA df
  token_TA = TOKEN_df.copy()

  # Calculate SMAs and add them to the DataFrame
  for t in timeperiods:
      ema = TA.SMA(token_TA, t)
      atr = TA.ATR(token_TA, t)
      adx = TA.ADX(token_TA, t)
      rsi = TA.RSI(token_TA, t)
      hma = TA.HMA(token_TA, t)
      vama = TA.VAMA(token_TA, t)

      token_TA[f'EMA_{t}'] = ema
      token_TA[f'ATR_{t}'] = atr
      # token_TA[f'ADX_{t}'] = adx
      # token_TA[f'RSI_{t}'] = rsi

  # Adding Chaikin Indicator 
  token_TA['CHAIKIN'] = TA.CHAIKIN(TOKEN_df)

  # # Adding Bollinger Bands
  # token_TA[['BB_UPPER','BB_MED','BB_LOWER']] =TA.BBANDS(TOKEN_df) 

  # Concatinate in the new TA features
  TOKEN_df = pd.concat([TOKEN_df,token_TA], axis=1) 

  return TOKEN_df
```

```python
# TEST to see if the new FTT_DF columns are larger
ftt_df = create_df_with_technical_analysis('FTT')
ftt_df
```

```python
# Token list to iterate through and create dfs
token_list = ['FTT','BNB','HT','OKB','KCS','KCS']

# Prints shape and quality to verify before downloading
for token in token_list:
  token_df = create_df_with_technical_analysis(token)
  token_df = token_df.dropna()
  print(token_df.shape)
  print(token_df.isnull().mean()*100)
  file = token_df.to_csv(f'{token}_dataset_with_TA.csv')
  files.download(f'{token}_dataset_with_TA.csv')
```
